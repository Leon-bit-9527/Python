```python
import random
import matplotlib.pyplot as plt
import numpy as np
def integral():
    n = 10000
    x_min,x_max = 0.0,1.0
    y_min,y_max = 0.0,1.0
    count = 0
    sum_y = 0
    for i in range(0,n):
        x = random.uniform(x_min,x_max)
        y = random.uniform(x_min,x_max)
        if pow((x-0.4),4)> y:
            count +=1
            plt.scatter(x, y)
        sum_y += pow((x-0.4),4)
    interal_value = count/float(n)
    interal_value2 = 1*sum_y/float(n)
    print("蒙特卡洛方法算出面积为：",interal_value)
    print("蒙特卡洛积分方法算出面积为：",interal_value2)
integral()
x = np.linspace(0, 1, 100)
plt.plot(x,pow((x-0.4),4))
```
